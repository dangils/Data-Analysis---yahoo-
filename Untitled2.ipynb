```python
import numpy as np

키 = np.array([170,180,160,165,158,176,182,172]).reshape((-1,1))
# .reshape((-1,1)) : 리스트를 2차원으로 만들어 주는 문법
몸무게 = [75,81,59,70,55,78,84,72] 

import matplotlib.pyplot as plt
plt.scatter(키,몸무게)
plt.show()

# LinearRegression 라이브러리로 OLS 충족하는 선 그리기
from sklearn.linear_model import LinearRegression

#하나의 데이터를 다시 리스트로 싸매서 fit에 전달
#numpy라이브러리로 데이터 변환
model = LinearRegression().fit(키,몸무게)

print(model.score(키,몸무게))
#R값 : 1과 가까우면 매우 밀접한 관련이 있다

#OLS 선 (y= ax+b)
print(model.intercept_) # a값

print(model.coef_) # b값
# -> y= 1.1x -118
# 몸무게 추정 가능 키가 170일때 69 = 187-118

a = model.predict([[170],[180]])
#x 데이터 입력가능
print(a)

plt.scatter(키,몸무게)
plt.plot(키,model.predict(키))
plt.show()

'''
회귀분석 사용:
1.데이터간 비례/반비례 여부 파악
2. 비례/반비례 할 경우 예측가능
3.두 데이터간의 관계 파악

# OLS를 충족하는선 : 최적의 선
OLS 알고리즘에 의해 선을 그림
경향성 및 예측 가능
Ordinary Least Squares 
'''
```

```python
#곡선으로 회귀분석

from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd 

df = pd.read_table('income.txt')
#tab으로 구분된 파일 읽어냄
df = df.dropna()
#빈 데이터 지우기
#scatter plot으로 파악, 두 데이터 간의 관계파악
'''
나이를 할면 소득을 추측할 수 있을까?
-> 곡선을 그려 소득을 추측 y=ax+bx^2+c
-> OLS를 만족하는 a,b,c를 찾음
- curve_fit()을 사용해 곡선 그리기
'''
plt.scatter( df['age'], df['income'] )
plt.show()

df

# curve_fit(함수,x,y)
def 함수(x,a,b,c):
  return a*x + b*x**2 + c
  #ax + bx^2 +c

opt, cov = curve_fit(함수, df['age'],df['income'])
#curve_fit() 실행하면 그 자리에 데이터 2종 남음
#a,b값을 추정해줌 ,opt=a,b,c 값 cov = 공분산 

print(opt)
a,b,c = opt #opt의 리스트 값들을 abc에 각각 담음

#찾은 곡선을 matplotlib으로 그려보자 
# y= 73x -8x^2 - 7

x = [1,2,3,4,5,6]

import numpy as np 
x = np.array([1,2,3,4,5,6])
# 리스트를 numpy array로 바꿔줌 , 행렬 쓸일이 있을때 array로 다룸
plt.scatter( df['age'], df['income'] )
plt.plot(x, 함수(x,a,b,c))
plt.show()
```

```python
import statsmodels.api as sm

'''
statsmodels로 Polynomial 분석
추정식 y = ax + bx^2 + c
x = [[첫놈나이, 첫놈나이^2,1]...]

const - 상수항 (1과 가깝거나 0.05 이상이면 관련이 없는 변수 )
이럴 경우 const를 없애면 좀더 적합한 모델

과적합 (overfitting)
-> 기존 데이터에만 적합 - 실용성이 없음 
'''

#column_stack : 원하는 리스트를 여러개 넣을 수 있음
x = np.column_stack([ df['age'], df['age']**2])
model = sm.OLS(df['income'], x).fit()
print(model.summary())
```
